```python
import spacy
nlp = spacy.load("en")
```

```python
import nltk
from nltk.corpus import stopwords
stopwords = stopwords.words('english')
```

```python
from nltk.tokenize import TreebankWordTokenizer
sentence_tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')
treebank_tokenizer = TreebankWordTokenizer()
```

### Read in data and toxic words

```python
import pandas as pd
jigsaw = pd.read_csv("/Users/ronnie/Downloads/6450NLP/jigsaw-unintended-bias-in-toxicity-classification/train.csv")
```

```python
dirty = jigsaw[jigsaw["target"]>=0.7]["comment_text"]
```

```python
toxicwords  = pd.read_csv("/Users/ronnie/Desktop/NLP Final/toxic50w.csv")
toxicwords
```

### Toxic words ralated phrases

```python
contexts = []
for comment in dirty: 
    analyzed = nlp(comment)
    lemmas = [token.lemma_.lower() for token in analyzed if not token.is_punct]
    intersect = set(lemmas).intersection(toxicwords)
    if len(intersect) > 0:
        context_sentences = [token.sent for token in analyzed if token.lemma_.lower() in toxicwords]
        contexts.extend(context_sentences)
contexts
```

```python
possible_toxic_phrases = []
for sentence in contexts:
    sentence = [token.lemma_.lower() for token in sentence if not token.is_punct | token.is_space]
    for i, token in enumerate(sentence):
        if token in toxicwords:
            
            phrase_1 = tuple(sentence[i-2:i+1])
            phrase_2 = tuple(sentence[i:i+3])
            if len(phrase_1)>1:
                possible_toxic_phrases.append(phrase_1)
            if len(phrase_2)>1:   
                possible_toxic_phrases.append(phrase_2)
possible_toxic_phrases
```

```python
phrase_dic = {}
for i in set(possible_toxic_phrases):
    if possible_toxic_phrases.count(i) >= 10:
        phrase_dic.update({i:possible_toxic_phrases.count(i)})   
phrase_dic
```

```python
sorted_dic = sorted(phrase_dic.items(), key = lambda item:item[1], reverse = True)
sorted_dic
```

### Nontoxic words based phrases

```python
puncs = '''!”“—’''[]{};:'’"\,<>./?@#$%^&*_~()--...``'''
```

```python
import string
cleaned_comments = []
for comment in dirty: 
    tokens = treebank_tokenizer.tokenize(comment)
    cleaned_comment = [token.strip('.').strip("….") for token in tokens if token not in toxicwords and (token not in puncs) and (token.lower() not in stopwords)]
    cleaned_comment = [word for word in cleaned_comment if word[0] !="'" and len(word)<15]
    cleaned_comments.append(cleaned_comment)
    
cleaned_comments
```

```python
possible_toxic_phrases = []
for comment in cleaned_comments:
    for i in range(len(comment)-2):
        phrase = tuple(comment[i:i+3]) 
        possible_toxic_phrases.append(phrase)
        
possible_toxic_phrases
```
